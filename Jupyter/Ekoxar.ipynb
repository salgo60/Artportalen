### Test API iNaturalist

```python
import requests
import matplotlib
import pandas
from urllib.parse import urlencode
import json

# Helpful site: https://api.inaturalist.org/v1/docs/#!/Observations/get_observations
# look for the taxon ids and place ids in the urls on the web site. 
# 
# https://www.inaturalist.org/observations?verifiable=true&taxon_id=61749&place_id=7599&preferred_place_id=7599&locale=sv

q = {
    "taxon_id": 61749, # Ekoze
    "d1": "2017-05-01",
    "d2": "2020-10-30",
    "order": "desc",
    "order_by": "created_at",
    "place_id": 7599, # Sverige
    "per_page": 200,
}
url = "https://api.inaturalist.org/v1/observations?"

iNat = pandas.DataFrame()
```

```python
r = requests.get(url, params = q)
results = r.json()["results"]
iNat = pandas.DataFrame(r.json()["results"])
# Print columns available
iNat.columns
```
